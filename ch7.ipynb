# 푸리에 계수로 톱니 파형 만들기

$$ f(t) = a_0 +  \sum_{n=1}^\infty  (a_n cos n wt + b_n sin n wt)$$

$$ f(t) = \sum_{n=1}^\infty \{-\frac{2}{n}(-1)^n \cdot sin nt\}$$

```python
import numpy as np
import matplotlib.pyplot as plt

# 벡터화된 푸리에 급수 함수
def saw_signal(t, N):
    n = np.arange(1, N + 1).reshape(-1, 1)  # (N, 1) 형태로 변형하여 브로드캐스팅
    terms = (-2 / n) * ((-1) ** n) * np.sin(n * t)  # (N, len(t)) 형태의 배열
    result = np.sum(terms, axis=0)  # n에 대해 합산 (각 t에 대한 결과)
    return result

# 시간 범위 및 고조파 수 설정
t = np.linspace(-10, 10, 1000)  # 시간축
N = 10000  # 고조파 수

# 함수 값 계산
signal = saw_signal(t, N)

# 그래프 그리기
plt.figure(figsize=(10, 5))
plt.plot(t, signal, label=f'Fourier Series Approximation (N={N})')
plt.title('Fourier Series Approximation of the Signal (Vectorized)')
plt.xlabel('Time (t)')
plt.ylabel('f(t)')
plt.grid(True)
plt.legend()
plt.show()


```
