# 매클로린 급수
$$ f(x) = \sum_{n=0}^\infty \frac{1}{n!}f^{(n)}(0)x^n $$

##  $f(x) = \sin (x) $ 를 매클로린 전개한다.

$ f(x) = \sin (x) $ \
$ f'(x) = \cos (x) $ \
$ f''(x) = -\sin (x) $\
$ f'''(x) = -\cos (x)$\
$ f''''(x) = \sin (x)$

$f(0) = \sin 0 = 0$\
$f'(0) = \cos 0 = 1$\
$f''(0) = -\sin 0 = 0 $\
$f'''(0) = -\cos 0 = -1 $\
$f''''(0) = \sin 0 = 0 $

$a_0 = \frac{1}{0!}f(0) = \frac{1}{0!}\cdot0 = 0 $\
$a_1 = \frac{1}{1!}f'(0) = \frac{1}{1!}\cdot1 = 1 $\
$a_2 = \frac{1}{2!}f''(0) = \frac{1}{2!}\cdot0 = 0 $ \
$a_3 = \frac{1}{3!}f'''(0) = \frac{1}{3!}\cdot -1 = -\frac{1}{3!} $\
$a_4 = \frac{1}{4!}f''''(0) = \frac{1}{4!}\cdot 0 = 0 $

$\sin x = 0 + x + 0x^2 -\frac{1}{3!}x^3 + 0x^4 + \frac{1}{5!}x^5 + 0 x^6 - \frac{1}{7!}x^7 + 0x^8 + ... $

```python
import numpy as np
import matplotlib.pyplot as plt
import math

fig = plt.figure()

x = np.linspace(-2 * np.pi,2 * np.pi,1000)
```

```python
y = 0 + x

plt.subplot(411)
plt.plot(x,y)
plt.xlim(-10,10)

y = y - ((1/math.factorial(3)) * (x ** 3))
plt.subplot(412)
plt.plot(x,y)
plt.xlim(-10,10)

y = y + ((1/math.factorial(5)) * (x ** 5)) - ((1 / math.factorial(7)) * (x ** 7)) +  ((1 / math.factorial(9)) * (x ** 9))
plt.subplot(413)
plt.plot(x,y)
plt.xlim(-10,10)

y = y - ((1 / math.factorial(11)) * (x ** 11)) +((1 / math.factorial(13)) * (x ** 13)) - ((1 / math.factorial(15)) * (x ** 15)) +  ((1 / math.factorial(17)) * (x ** 17))
plt.subplot(414)
plt.plot(x,y)
plt.xlim(-10,10)

plt.show()
```
